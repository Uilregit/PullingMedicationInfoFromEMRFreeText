# Start splitting data into Tokens

```python
# start the tokenization process
import re

tokens = []
for paragraph in data:
    tk = re.split(" |,|-|:", paragraph.replace(".",""))
    tk = [i.lower() for i in tk if len(i) > 0]
    tokens += [tk]
    
print (tokens)
```

```python
# getting to see if the words are nouns, adverbs, adjectives
from nltk.corpus import wordnet as wn

# WordNet POS tags are: NOUN = 'n', ADJ = 's', VERB = 'v', ADV = 'r', ADJ_SAT = 'a'
# Descriptions (c) https://web.stanford.edu/~jurafsky/slp3/10.pdf
tag_map = {
        'CC':None, # coordin. conjunction (and, but, or)  
        'CD':wn.NOUN, # cardinal number (one, two)             
        'DT':None, # determiner (a, the)                    
        'EX':wn.ADV, # existential ‘there’ (there)           
        'FW':None, # foreign word (mea culpa)             
        'IN':wn.ADV, # preposition/sub-conj (of, in, by)   
        'JJ':wn.ADJ, # adjective (yellow)                  
        'JJR':wn.ADJ, # adj., comparative (bigger)          
        'JJS':wn.ADJ, # adj., superlative (wildest)           
        'LS':None, # list item marker (1, 2, One)          
        'MD':None, # modal (can, should)                    
        'NN':wn.NOUN, # noun, sing. or mass (llama)          
        'NNS':wn.NOUN, # noun, plural (llamas)                  
        'NNP':wn.NOUN, # proper noun, sing. (IBM)              
        'NNPS':wn.NOUN, # proper noun, plural (Carolinas)
        'PDT':wn.ADJ, # predeterminer (all, both)            
        'POS':None, # possessive ending (’s )               
        'PRP':None, # personal pronoun (I, you, he)     
        'PRP$':None, # possessive pronoun (your, one’s)    
        'RB':wn.ADV, # adverb (quickly, never)            
        'RBR':wn.ADV, # adverb, comparative (faster)        
        'RBS':wn.ADV, # adverb, superlative (fastest)     
        'RP':wn.ADJ, # particle (up, off)
        'SYM':None, # symbol (+,%, &)
        'TO':None, # “to” (to)
        'UH':None, # interjection (ah, oops)
        'VB':wn.VERB, # verb base form (eat)
        'VBD':wn.VERB, # verb past tense (ate)
        'VBG':wn.VERB, # verb gerund (eating)
        'VBN':wn.VERB, # verb past participle (eaten)
        'VBP':wn.VERB, # verb non-3sg pres (eat)
        'VBZ':wn.VERB, # verb 3sg pres (eats)
        'WDT':None, # wh-determiner (which, that)
        'WP':None, # wh-pronoun (what, who)
        'WP$':None, # possessive (wh- whose)
        'WRB':None, # wh-adverb (how, where)
}

def get_wordnet_pos(tag):
    try:
        if tag_map[tag] != None:
            return tag_map[tag]
    except:
        return wn.NOUN
    return wn.NOUN
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

def get_lemmated_toekens(tokens):
    stop_words = stopwords.words("english")
    lem = WordNetLemmatizer()
    
    tokens = [i.lower() for i in tokens if i != ""]
    tokens = nltk.pos_tag(tokens)
    
    outputTokens = []
    for i in range(len(tokens)):
        if (tokens[i][1] == "IN"):            #intentially leaves injunctions in the tokens list
            outputTokens += [tokens[i][0]]
        elif tokens[i][0] not in stop_words and tokens[i][0].isalpha():
            outputTokens += [lem.lemmatize(tokens[i][0], get_wordnet_pos(tokens[i][1]))]   #converts all other words into the base form
    return outputTokens
```

# Get Lemmantized and cleaned Tokens

```python
# gets every word into an array, lowercase, stemming and lemmitization
cleanedTokens = []
for paragraph in words:
    cleanedTokens += [get_lemmated_toekens(paragraph)]

print (cleanedTokens[0])
```

# Find locations of labeled medications in the lemmatized paragraph

```python
import Levenshtein as lv
def checkIfLabelsInParagraph (labels, paragraph):
    if labels == [""]:
        return [None]
    else:
        labels = [i.lower() for i in labels]
    output = []
    for item in labels:
        if item in paragraph:   #if exact match then return 0
            output += [paragraph.index(item)]
        else:                   #if not an exact match
            levenDist = []
            wordsInLabels = len(item.split(" "))  #checks for multiword labels
            for i in range(len(paragraph)-wordsInLabels+1):
                levenDist += [lv.distance(item, " ".join(paragraph[i:i+wordsInLabels]))]    #checks for misspellings
            #print (levenDist)
            output += [levenDist.index(min(levenDist))]
    return output
```

```python
labels1 = pd.read_csv("Test1+2GroundTruth.csv", header = None, index_col = 0)
labels2 = pd.read_csv("Test3GroundTruth.csv", header = None, index_col = 0)
labels3 = pd.read_csv("mt_chart_labelled.csv", header = None, index_col = 0)
labels4 = pd.read_csv("mt_discharge_labelled.csv", header = None, index_col = 0)

labels = labels1.append(labels2).append(labels3).append(labels4)
labels.columns = ["Meds"]
labels["Meds"] = labels["Meds"].str.lower()
labels = labels.reset_index(drop = True)

labelsTokens = [[""] if type(i) == float else i.split(", ") for i in labels["Meds"]]

location = []
for i in range (len(cleanedTokens)):
    location += [checkIfLabelsInParagraph(labelsTokens[i], cleanedTokens[i])]
locationDf = pd.DataFrame(location)
locationDf.head(10)
```

# Find Neighbour and AntiNeighbour tokens

```python
def getNeighbours (neighbourhood = 5, tokenSize = 1):
    neighbourVocab = []

    for i in range(len(location)):
        if location[i][0]!= None:
            for j in location[i]:
                for k in range(neighbourhood):
                    words = [" ".join(cleanedTokens[i][j-k-tokenSize:j-k])] + [" ".join(cleanedTokens[i][j+k+1:j+k+tokenSize+1])]
                    #words = cleanedTokens[i][j-neighbourhood:j] + cleanedTokens[i][j+1:j+neighbourhood+1]
                    neighbourVocab += words
    neighbourVocabDf = pd.DataFrame(neighbourVocab)
    neighbourVocabDf = neighbourVocabDf[neighbourVocabDf[0]!=""]
    return neighbourVocabDf[0].value_counts()
```

```python
import copy

def GetAntiNeighbours (neighbours, neighbourhood = 5):
    antiNeighbourVocab = []
    for i in range(len(cleanedTokens)):
        for neighbour in neighbours:
            tempTokens = copy.deepcopy(cleanedTokens[i])
            #print(neighbour)
            while neighbour in tempTokens:
                neighbourIndex = tempTokens.index(neighbour)
                #print (neighbourIndex)
                #print (location[i])
                if (location[i]!= [None]):
                    distanceToMed = [abs(neighbourIndex-j) for j in location[i]]
                else:
                    distanceToMed = [999999]
                
                #print (min(distanceToMed))
                if min(distanceToMed) > neighbourhood:
                    antiNeighbourVocab += [tempTokens[neighbourIndex-neighbourhood:neighbourIndex-1]+tempTokens[neighbourIndex+1:neighbourIndex+neighbourhood]]
                tempTokens.pop(neighbourIndex)
    antiNeighbourVocabDf = pd.DataFrame(antiNeighbourVocab)
    print (antiNeighbourVocabDf)
    antiNeighbourVocabDf = antiNeighbourVocabDf[antiNeighbourVocabDf[0]!=""]
    return antiNeighbourVocabDf[0].value_counts()
```
